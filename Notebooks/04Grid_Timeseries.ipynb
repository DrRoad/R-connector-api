Input a startdate, an enddate and the time interval, all as datetime-objects. The interval tells you, if you get the data in hourly steps, daily steps or every five minutes in between the startdate and the enddate. Example:

```r
time_zone = "Europe/Berlin"
startdate = ISOdatetime(year = as.integer(strftime(today(),'%Y')), 
                        month = as.integer(strftime(today(),'%m')), 
                        day = as.integer(strftime(today(),'%d')), 
                        hour = 00, min = 00, sec = 00, tz = "UTC")
enddate = ISOdatetime(year = as.integer(strftime(today(),'%Y')), 
                      month = as.integer(strftime(today(),'%m')), 
                      day = as.integer(strftime(today(),'%d'))+1, 
                      hour = 00, min = 00, sec = 00, tz = "UTC")
interval = "PT12H"
```

Set parameter and location.

```r
parameters = "t_2m:C"
coordinate = "48.5,5.5_45.5,11.5:600x500"
```

There is the option to plot the grids and save them to an arbitrary working directory.

```r
timeseries_grid <- timeseries_grid(startdate, enddate, interval, parameter, coordinate, plot = TRUE)
```
